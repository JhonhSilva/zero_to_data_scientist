# About Dataset

<p> This dataset focuses on analyzing daily water intake and hydration levels of individuals based on a combination of demographic, lifestyle, and environmental factors. Each row represents a single individual and captures how variables such as age, gender, body weight, physical activity level, and weather conditions influence daily water consumption and overall hydration status. </p>

<p> The dataset is designed to support exploratory data analysis (EDA), health behavior studies, and machine learning classification tasks, where the goal is to predict whether an individual is properly hydrated (Good) or at risk of dehydration (Poor). It is well-suited for beginners and intermediate learners working on health analytics, lifestyle prediction models, and feature impact analysis. </p>

### This dataset can be used for:

* Hydration level prediction
* Lifestyle and wellness analysis
* Feature importance and correlation studies
* Binary classification models (Good vs Poor hydration)

### About this file

* Number of Rows: 30000
* Number of Columns: 7
* Target Variable: Hydration Level
